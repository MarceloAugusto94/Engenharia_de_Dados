```python
## FORMA MAIS UTILIZADA

from pyspark.sql import SparkSession

spark = (SparkSession.builder 
    .master("local[3]") 
    .appName("Estudando-DataFrame") 
    .getOrCreate())

print(spark)
```

```python
from pyspark import SparkConf

conf = SparkConf() \
    .setMaster("local") \
    .setAppName("Estudando-DataFrame")

spark = SparkSession.builder.config(conf = conf).getOrCreate()
```

```python
df = (spark
      .read
      .format("csv")
      .option("header", "true")
      .option("inferSchema", "true")
      .option("delimiter", ";")
      .load("dataset/arquivo_geral.csv"))
```

```python
df.show(10)
```

```python
# Visualizar o Schema das minhas colunas

df.printSchema()
```

```python
df.select("regiao", "estado", "casosNovos").show(10)
```

```python
import pyspark.sql.functions as f
```

```python
# Funções como string

print(dir("regiao"))
```

```python
# Funções como coluna

print(dir(f.col("regiao")))
```

```python
df.select(
    f.col("regiao"), 
    f.col("estado"), 
    f.col("casosNovos")).filter(f.col("regiao") == "Norte").show(10)


# 2 forma de fazer
df.select(
    f.col("regiao"), 
    f.col("estado"), 
    f.col("casosNovos")).filter(df.regiao == "Norte").show(10)

# 3 forma de fazer
df.select(
    f.col("regiao"), 
    f.col("estado"), 
    f.col("casosNovos")).filter("regiao = 'Norte'").show(10)
```

```python
# 4 forma de fazer
filtro = f.col("regiao") == "Sul"

df.select(
    f.col("regiao"), 
    f.col("estado"), 
    f.col("casosNovos")).filter(filtro).show(10)
```

```python
col_select = ["regiao", "casosNovos"]
df.select(*col_select).show(100)
```

```python
# 1 Forma
df.filter("regiao = 'Norte' and estado = 'AM'").show(10)

# 2 Forma
df.filter("regiao = 'Norte'").filter("estado = 'AM'").show(10)

# 3 Forma
df.filter((f.col("regiao") == "Norte") & (f.col("Estado") == "AM")).show(10)

# 4 Forma
df.filter((f.col("regiao") == "Norte") | (f.col("Estado") == "AM")).show(10)

# 5 Forma
df.where((f.col("regiao") == "Norte") | (f.col("Estado") == "AM")).show(10)
```

```python
# 1 Forma
df.filter(f.col("regiao") == "Norte").filter("Estado like 'A%'").show(100)
```

```python
df.filter("regiao in ('Norte', 'Sul')").show(1000)

# 2 Forma
list_region = ["Norte", "Sul"]

df.filter(f.col("regiao").isin(list_region)).show(1000)
```

```python
df.filter(f.col("regiao").startswith("C")).show(100)

# 2 Forma
df.filter(f.col("regiao").like("C%")).show(100)
```
