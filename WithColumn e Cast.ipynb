```python
df = (spark
     .read
     .format("csv")
     .option("header", "True")
     .option("inferSchema", "True")
     .option("Delimiter", ";")
     .load("dataset/arquivo_geral.csv"))

df.printSchema()
```

```python
import pyspark.sql.functions as f
```

```python
new_df = df.withColumn("soma_novos_e_acumulados", f.col("casosNovos") + f.col("casosAcumulados"))
```

```python
new_df.show(1000)
```

```python
new_df = new_df.withColumn("coluna_personalizada", f.lit("O valor da coluna"))
```

```python
new_df.show(100)
```

```python
new_df.printSchema()

new_df_cast = new_df.select(f.col("casosNovos").cast("string"))

new_df_cast.printSchema()
```

```python
new_df_cast = new_df.selectExpr("cast(casosNovos as string)")

new_df_cast.printSchema()
```
