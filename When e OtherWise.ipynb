```python
df = (spark
     .read
     .format("csv")
     .option("header", "True")
     .option("inferSchema", "True")
     .option("Delimiter", ";")
     .load("dataset/arquivo_geral.csv")
     )

import pyspark.sql.functions as f
```

```python
df.printSchema()

df.show(10)
```

```python
df_status = (df.withColumn("status", f.when(f.col("casosNovos") > 0, "Tem casos novos").otherwise("Sem casos novos")))

df_status.show(1000)
```

```python
df_saldo_status = (df.withColumn("status", f.when(f.col("casosNovos") > 0, 
                                 f.concat(df.casosNovos, f.lit(" casos novos"))).otherwise("Sem casos novos")))
df_saldo_status.show(1000)
```

```python
df_saldo_status = (df.withColumn("Dia", f.substring(f.col("data"), 9, 2)).
                   withColumn("Ms", f.substring(f.col("data"), 6, 2)).
                   withColumn("Ano", f.substring(f.col("data"), 1, 4)))
                   
                   
df_saldo_status.show(10)
```

```python
df_saldo_status.printSchema()
```

```python
df_saldo_status = (df.withColumn("Dia", f.substring(f.col("data"), 9, 2).cast("integer")).
                   withColumn("Ms", f.substring(f.col("data"), 6, 2).cast("integer")).
                   withColumn("Ano", f.substring(f.col("data"), 1, 4).cast("integer")))
                   
                   
df_saldo_status.show(10)
```

```python
df_saldo_status.printSchema()
```
